### For Loops

````python
# for keyword, counter/iterator, in keyword, object

for i in range(10):
    print(i)

print("\n")
name = "Marc Engle"
for letter in name:
    if letter == " ":
        print("This is a space!")
    else:
        print(letter)

    
````

##### Using 'in' keyword

````python
#in is what accesses the items in an iterable
for num in range(15):
    print(num**2)
````

##### Continue Statement

````python
#continue

for i in range(5):
    print(i)
for i in range(10):
    if i == 5 or i == 6:
        continue
    print(i)
````

##### Break Statement

````python
#will break out of current loop
for i in range(25):
    if i == 23:
        break
    print(i)
````

##### Pass Statement

##### Double For Loops

````python
for i in range(5):
    print("i = ", i)
    for j in range(5):
        print("j = ", j)
````

### While Loops

````python
#while keyword, condition statement
#while loop will run based on the condition statement

counter = 0
while counter < 11:
    counter += 1
    print(counter)
    
counter1 = 1
while counter1 < 100:
    print(counter1)
    counter1 += 2

````

##### looping 'While True'

````python
# true as a condition for while loop to run
game_over = True
counter = 0
while game_over:
    counter += 1
    print("Infinite Loops")
    if counter == 10:
        game_over = False
        
````

##### While & For Loops Used Together

````python
j = 0

while j < 10:
    print("\n WHile loop iteration: " + str(j))
    for i in range(2):
        print ("For loop iteration: " + str(1))
        
    j += 1
````

##### The Range Function

````python
# range(start, stop, step) - stop is required and start defaults to 0 if none is given
for i in range(10): #stop
    print(i)
print("\n")
for i in range(2, 10): #start and stop
    print(i)
print("\n")  
for i in range(1, 20, 3): #start, stop, step
    print(i)
````

##### Declaring Lists

##### Indexing a List

````python
# [start: stop: step] --> last item [-1]
names = ["Ryan", "Alex", "Kate", "Saraa", "Sam", "Gus"]

#single index
print(names[1])
name = names[0]
print(name)
print(names[-1]) #-1 will always be the last item in a list

print("\n")
#print starting at index 1 and going to the end of the list
print(names[1:])
new_names = names[2:]
print(new_names)

print("\n")
#print starting at the beginning of the list and going to a specified index (just like range it is non-inclusive)
print(names[:2])
print("\n")

#print starting at index 1 and stepping by 2
print(names[1::2])

#print starting at the back and go to the front of the list
print(names[::-1])
reverse_list = names[::-1]
print(reverse_list)
````

### List Methods

##### .append()

````python
#always adds an element to the end of a list
#our_list.append(thing we're appending)
names = ["Ryan", "Alex", "Kate", "Saraa", "Sam", "Gus"]

names.append('Derrick')
print(names)

names.append("Aryan")
print(names)
````

##### .insert()

````python
#our_list.insert(index, value)
names = ["Ryan", "Alex", "Kate", "Saraa", "Sam", "Gus"]

names.insert(2, "Marc")
print(names)

names.insert(len(names), "Eduardo")
names.append("Eduardo")
print(names)
````

##### .pop()

````python
# default is the last value if no parameter is given
# pop returns the element that was removed in case you want to assign it into a variable

names = ["Ryan", "Alex", "Kate", "Saraa", "Sam", "Gus"]
names.pop(0)
print(names)

popped_friend = names.pop(2)
print(popped_friend)
print(names)

names.pop()
print(names)




````

##### .remove()

````python
# value to be removed, not an index
names = ['Bob', 'Billy', 'Berry', "Bob", "Bertha", 'Bob', "Bob", 'Bert']

for name in names:
    if name == "Bob":
        names.remove(name)

print(names)
````

````python
#Three ways to loop through a list. one is by index and the other is by value, using the "in" keyword
names = ["Han Solo", "Leia Organa", "Luke Skywalker", "Lando Calrissian", "Mace Windu"]


#Looping by value
for name in names:
    print(name)

print("\n")
#Looping by index with range(len)
# print(len(names))
# for i in range(5):
#     print(i)

for i in range(len(names)):
    print(names[i])
    
print("\n")    
#enumerate
for i, name in enumerate(names):
    print(i, name)

# print(list(enumerate(names)))
````

### List Comprehensions <br>
<p>Creating a quickly generated list to work with<br>*result*  = [*transform*    *iteration*         *filter*     ]</p>

##### In a list comprehension we have a few pieces:
1. The first is the counter/ variable - IN this the variable is x
2. then we have a transform for the variable
3. The finale part of a list comp is called the condition

```python
    [transform, iteration/variable, condition]
```

````python
nums = [2, 4, 5, 6, 7, 8, 9, 10]
nums_squared = []
for num in nums:
    nums_squared.append(num**2)
print(nums_squared)

nums_squared2 = [num**2 for num in nums]
print(nums_squared2)

nums_squared3 = [num**2 for num in nums if num%2==0]
print(nums_squared3)
````

````python
print([num**2 for num in nums if num%2==0])
````

There are a few benefits to using List comprehensions. The most obvious would be that we now have shorter code to work with instead of using 3+ lines of code in the for loop variant.

Another is an added benefit to memory usage. Since the list's memory is allocated first before adding elements to it, we don't have to resize the list once we add elements to it.

Lastly, list comprehensions are considered the "pythonic" way to write code by the PEP8 standards (Python Style Guide)
